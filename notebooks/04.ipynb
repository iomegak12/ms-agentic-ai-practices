# Step 04: Using Semantic Functions from Files

This notebook demonstrates how to load semantic functions from external prompt files.

```python
import asyncio
from semantic_kernel import Kernel
from semantic_kernel.connectors.ai.open_ai import AzureChatCompletion
```

```python
async def main():
    kernel = Kernel()
    chat_completion = AzureChatCompletion()
    kernel.add_service(chat_completion)

    semantic_functions = kernel.add_plugin(parent_directory=".", plugin_name="prompts")
    activities_function = semantic_functions["activities"]

    response = await kernel.invoke(activities_function, city="Oslo", time="1PM")

    print(response)
```

```python
# Run the main function
await main()
```
