# Step 01: Basic Semantic Kernel with Azure OpenAI

This notebook demonstrates how to set up a basic Semantic Kernel with Azure OpenAI chat completion.

```python
import asyncio
from semantic_kernel import Kernel
from semantic_kernel.connectors.ai.open_ai import AzureChatCompletion
```

```python
async def main():
    kernel = Kernel()
    chat_completion = AzureChatCompletion()
    kernel.add_service(chat_completion)

    prompt = "What can I do near where I live at the moment?"

    response = await kernel.invoke_prompt(prompt=prompt)
    print(response)
```

```python
# Run the main function
await main()
```
